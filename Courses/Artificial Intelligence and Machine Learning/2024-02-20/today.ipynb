```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
raw_df = pd.read_csv("https://raw.githubusercontent.com/datagy/data/main/stocks.csv")
raw_df.plot(kind='line')
plt.title('MSFT price over time')
plt.ylabel('Price of MSFT')
plt.xlabel('Time')
plt.show()
```
