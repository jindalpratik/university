```python
# EXAMPLE (SEQUENCE)

# Sequence is the ordered collection of similar or different data types
```

```python
# Strings are arrays of bytes representing Unicode characters. 
# It is a collection of one or more characters put in a single quote, double-quote or triple quote.
```

```python
# Example Strings (Sequence)

# String with single quotes
String1 = 'CSE Student'
print("Single Quotes") 
print(String1) 
    
# String with double Quotes 
String2 = "CSE Student's"
print("Double Quotes") 
print(String2) 
print(type(String2))
    
# String with triple Quotes 
s = '''CSE Student's doing "well"s'''        
print("\n\nTriple Quotes") 
print(s) 
print(type(s))

print(s[1])
print(s[5])
print(s[-3])
```

```python
# EXAMPLE (LISTS)

# Lists are a ordered collection of data.
# Lists are mutable
```

```python

f="vinay kukreja"
print(type(f))
e=[78,122,53,24,673]
print(e[3])

List = ["CSE", "Students", "Chitkara"] 
print(List)
print(List[1])  
print(List[2])
print(List[-3])

List[0]="ECE"   # MUTABLE
print(List)

List[1]="ME"
print(List)
```

```python
# Creating a Blank List 
List = [] 
print(List)

# Create List with String 
List = ['CSE STUDENT'] 
print(List) 
    
# List with multiple string values 
List = ["CSE", "Students", "Chitkara"] 
print(List)
print(List[1])  
print(List[2])
print(List[-3])
List[0]="BTECH CSE"   # MUTABLE
print(List)
    
# Nesting of Lists
List = [['CSE', 'Students'], ['Chitkara']] 
print(List) 
print(List[1])
print(List[0])
print(List[0][1])
```

```python
# EXAMPLE (TUPLE)

# Tuple is also an ordered collection of Python objects.
# Tuple are immutable
```

```python
d=(1,2,3,123.34,"Python",True)
print(type(d))
```

```python
# Creating an empty tuple 
Tuple1 = () 
print(Tuple1) 
    
# Create Tuple with Strings 
Tuple2 = ('CSE', 'Students') 
print(Tuple2) 
    
# Create Tuple with the help of list 
list1 = [2, 6, 10, 15] 
print(type(tuple(list1))) 


  
# Create Tuple with the built-in function tuple 
Tuple3 = tuple('CSE Students') 
print(Tuple3) 
  
# Create Tuple with nested tuples 
Tuple4 = (2, 4, 8) 
Tuple5 = ('CSE', 'Students') 
Tuple6 = (Tuple4, Tuple5) 
print(Tuple6) 
```

```python
tuple1 = (12, 22, 23, 44, 53)
print(tuple1)
print(tuple1[4])
print(tuple1[-4])
tuple1[0]=456    # IMMUATBLE
```

```python
# EXAMPLE (SET)

# It is an unordered collection of data type that is iterable, mutable and has no duplicate elements 
```

```python
f={1,2,3,4,2,5,4}
print(f)
print(type(f))
```

```python
# Creating a Set 
set1 = set() 
print(set1) 
    
# Create Set with a String 
set2 = set("CSE STudents") 
print(set2) 
  
# Create Set with a List 
set3 = set(["CSE", "Students", "Chitkara"]) 
print(set3) 
  
# Create Set with mixed type of values 
set4 = set([4, 12, 'CSE', 18, 'Students', 12, 'CSE']) 
print(set4) 
```

```python
# EXAMPLE DICTIONARY

# Dictionary in Python is an unordered collection of data values, used to store data values
# Dictionary holds key:value
```

```python
g={0:'win',1:'great',2:'home'}
print(type(g))
print(g)
```

```python
# Empty Dictionary
Dict1 = {}
print(Dict1)
 
# Dictionary with dict() method
Dict2 = dict({1: 'Geeks', 2: 'For', 3:'Geeks'})
print(Dict2)
 
# Dictionary with each item as a Pair
Dict3 = dict([('name', 'Vinay'), (7, 'Kukreja')])
print(Dict3)


print(Dict3[7])
print(Dict3['name'])
```
